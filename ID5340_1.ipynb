```python
# linear regression model

data=pd.read_csv("LR_data.csv")
data
```

```python
plt.scatter(data["x"],data["y"])
plt.show()
```

```python
def linear_1d(w_vec,x_vec=data["x"]):
    w0=w_vec[0]
    w1=w_vec[1]

    y_pred=w0+w1*x_vec
    return y_pred
```

```python
w_arb=[6,-2]
y_arbi_pred=linear_1d(w_arb,x_vec=data["x"])
```

```python
plt.plot(data["x"],y_arbi_pred,color="black")
plt.scatter(data["x"],data["y"],color="red")
plt.show()
```

```python
arbi_error=np.mean((data["y"]-y_arbi_pred)**2)
```

```python
print("Mean Square Error: %.2f" %arbi_error)
```

```python
x_avg=np.mean(data["x"])
y_avg=np.mean(data["y"])

w_optimal_1=np.sum((data["x"]-x_avg)*(data["y"]-y_avg))/np.sum((data["x"]-x_avg)**2)
w_optimal_0=y_avg-w_optimal_1*x_avg
```

```python
print("W0: %.2f" %w_optimal_0)
print("W1: %.2f" %w_optimal_1)
```

```python
w_vec_optimal=[w_optimal_0,w_optimal_1]
y_pred_optimal=linear_1d(w_vec_optimal,data["x"])
```

```python
plt.plot(data["x"],y_pred_optimal,color="black")
plt.scatter(data["x"],data[["y"]],color="red")
plt.plot(data["x"],y_arbi_pred,color="blue")

plt.show()
```

```python
optimal_error=np.mean((data["y"]-y_pred_optimal)**2)
```

```python
print("Mean Square Error: %.2f" %optimal_error)
```

```python
# make predictions for new data

x_new=np.array([0.1,0.8])
y_new=w_optimal_1*x_new+w_optimal_0
```

```python
plt.plot(data["x"],y_pred_optimal,color="black",label="optimal_prediction")
plt.scatter(data["x"],data[["y"]],color="orange",label="data")
plt.plot(data["x"],y_arbi_pred,color="blue",label="arbitary prediction")
plt.scatter(x_new,y_new,color="green",label="new_data")
plt.legend()
plt.show()
```

```python
import torch

x=torch.rand(10,1,dtype=torch.float)
y=torch.rand(10,1,dtype=torch.float)

w=0.5
b=1

learning_rate=0.001
error=1e-2

def grad_w(y,y_pred):
    return -2*torch.mean((y-y_pred)*x)

def grad_b(y,y_pred):
    return -2*torch.mean((y-y_pred))

while (1):
    y_pred=w*x+b
    loss=torch.mean((y-y_pred)**2)

    w=w-grad_w(y,y_pred)*learning_rate
    b=b-grad_b(y,y_pred)*learning_rate
    
    print(loss)

    
    if loss<error:
        break
```
